#### 3.1.4 What is the characteristic of our regular customer?

To answer this question, we need to define `regular customer`. There's no strict definition of regular customer, however, behind this question is that we want to know what is the characterics of our customer that may indicate that they will visit regularly.  
Thus, we'll define regular customer by:
- `Frequency of visit`: including the customer in 'Weekly' and 'Daily' category, and
- `Spending`: Based on the latest Starbuck Reward program, which is spend-based loyalty program, we know that Starbuck valued customer of higher spending. Thus, we'll  include 'Monthly' customer that spend higher than mode of the average spend.

```python
## Determine the mode of average spend per visit.
df["avg_spend_per_visit"].value_counts()
```

```python
## Determining customer that spend more than mode of average spend per visit.
spend_1 = df["avg_spend_per_visit"]=="Around RM20 - RM40"
spend_2 = df["avg_spend_per_visit"]=="More than RM40"
freq_monthly = df["freq_of_visit"]=="Monthly"

monthly_spend = freq_monthly & (spend_1 | spend_2)

## Determining the customer with frequency of visit of 'Weekly' and 'Daily'.
weekly = df["freq_of_visit"]=="Weekly"
daily = df["freq_of_visit"]=="Daily"

frequent = weekly | daily

## Defining our regular_customer.
regular_customer = df[monthly_spend | frequent]
regular_customer.head()
```

```python
regular_customer_demo = regular_customer[["gender","age_cat","status","annual_income","nearest_franchise"]]
regular_customer_demo.head()
```

```python
for i in regular_customer_demo:
    demo = (regular_customer_demo[i].value_counts(normalize=True))*100
    print(demo)
    print("----------------------------------")
```

We identified the characteristics of our regular customers as:
1. **Gender Distribution**: The majority of our regular customer are male, with a proportion of 65.52%, while female account for 34.48%.
2. **Age Group**: The largest age group among Starbuck regular customer is "From 20 to 29", representing 58.62%. The second-largest age group is "From 30 - 39", with 24.14%, followed by "40 and above" at 13.79%, and "Below 20" as 3.45%.
3. **Status**: The majority of Starbucks regular customers are employed (62.07%), followed by self-employed individuals (24.14%), students (10.34%), and housewives (3.45%).
4. **Annual Income**: The annual income distribution among Starbucks regular customers is evenly split, with 31.03% earning "Less than RM25,000", 31.03% earning "RM50,000 - RM100,000", 27.59% earning "RM25,000 - RM50,000", and 10.34% earning "More than RM150,000."
5. **Proximity to franchise**: A significant proportion of Starbucks regular customers (44.83%) are located "more than 3km" away from a Starbucks franchise, followed by those located "1km - 3km" away (31.03%), and "within 1km" (24.14%).

As summary, these insights suggest that Starbucks' regular customer base is predominantly young to middle-aged working professionals and students, with a moderate to high income level. Substantial portion of regular customers willing to travel longer distances to visit a Starbucks franchise, which may indicate a strong brand-loyalty.

#### 3.2.5 Does order option, visit duration, and type of frequently purchased item correlated with average spend per visit?

```python
## i. Order option and average spend per visit.
cust_behaviour[["order_option","avg_spend_per_visit"]].groupby("order_option").value_counts(normalize=True)*100
```

```python
order = ["I dont like coffee", "Never", "Dine in", "Take away", "Drive-thru"]
hue_order = ["Zero", "Less than RM20", "Around RM20 - RM40", "More than RM40"]
pastel = sns.color_palette("pastel")

g = sns.countplot(data=cust_behaviour,
            x="order_option",
            hue="avg_spend_per_visit",
            palette=pastel,
            order=order,
            hue_order=hue_order)

plt.title("Average spend per visit by Order option", y=1.02)
plt.xlabel("Order option")
plt.ylabel("Count")

legend_title = "Average spend per visit"
g.legend_.set_title(legend_title)

plt.show()
```

For `dine in`, most customer(54.3%) prefer to spend less than RM20, while 32.6% of them prefer to spend around RM20 - RM 40. Only a small portion of the customer 6.5% would spend more than RM40 per visit.  
The last 6.5% of customer would not spend in dine in.  

`Take away` option follow the same pattern as dine in where most of the customer(55.1%) prefer to spend less than RM20, and customer spending around RM20 - RM 40 come in second place with 36.7%. Some the customer(6.12%) would not spend in take away option and only 1 person(2.04%) would spend more than RM 40.

The highest percentage(55%) of customer in `Drive-thru` category would spend around RM20 - RM40. The customer spending less than RM20 account for 30% of Drive-thru customer and last 15% represent customer spending More than RM 40.


```python
## ii. Visit duration and average spend per visit
cust_behaviour[["visit_duration", "avg_spend_per_visit"]].groupby("visit_duration").value_counts(normalize=True)*100
```

```python
order = ["Below 30 minutes", "Between 30 minutes to 1 hour", "Between 1 hour to 2 hours", "Between 2 hours to 3 hours", "More than 3 hours"]
hue_order = ["Zero", "Less than RM20", "Around RM20 - RM40", "More than RM40"]
pastel = sns.color_palette("pastel")

g = sns.countplot(data=cust_behaviour,
            x="visit_duration",
            hue="avg_spend_per_visit",
            palette=pastel,
            order=order,
            hue_order=hue_order)

plt.title("Average spend per visit by Visit duration", y=1.02)
plt.xlabel("Visit Duration")
plt.ylabel("Count")
plt.xticks(rotation = 90)

legend_title = "Average spend per visit"
g.legend_.set_title(legend_title)


plt.show()
```

Our data reveals an inverse relationship between visit duration and customer count. The highest concentration of customers falls within the `Below 30 minutes` category, followed by `Between 30 minutes to 1 hour`. This trend continues with a significant drop in customer count for the `Between 1 hour to 2 hours` category. Visits exceeding 2 hours (`Between 2 hours to 3 hours` and `More than 3 hours`) represent a minority of customers.

Average spending exhibits a similar pattern - where significant proportion of customer will spend less than RM20, followed by RM20 - RM40 - within the most frequent visit durations (`Below 30 minutes` and `Between 30 minutes to 1 hour`). However, the distinction in average spend between these categories is less pronounced compared to the decrease in customer count for longer visits.

In conclusion, while shorter visits are more prevalent, the data suggests a weak correlation between visit duration and average spending. This indicates that customers with longer visits are not necessarily spending significantly more per visit.
